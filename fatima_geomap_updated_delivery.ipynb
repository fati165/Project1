```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key

# Study data files
yelp_path = "data/Final_Yelp_Data.csv"

# Read the mouse data and the study results
yelpers = pd.read_csv(yelp_path)
```

```python
yelpers.head()
```

```python
#compare each type of food and review count
# grab categories, name, review count, and rating
df = pd.DataFrame(yelpers, columns = ['name', 'categories', 'review_count', 'rating', 'transactions', 'coordinates']) 

df.head()
```

```python
#geomap the restaurants with most reviews
#top 20 and low 20
#df = pd.DataFrame(df.row.str.split(' ',1).tolist(),
#                                 columns = ['coordinates'])
df['lat'], df['long'] = df['coordinates'].str.split(',', 1).str
df['lat']= df['lat'].str.replace("{'latitude':", "")
df['long'] = df['long'].str.replace("'longitude': ", "").str.replace("}", "")
df
df[['long', 'lat']]= df[['long', 'lat']].astype(float, errors= "ignore")
#df['long'].isna().sum()
df
```

```python
df= df.sort_values("review_count", ascending= False)
top20= df.head(20)
top20[['long', 'lat']]= top20[['long', 'lat']].astype(float)

```

```python
df= df.sort_values("review_count", ascending= False)
tail= df[df["review_count"]==1]
tail[['long', 'lat']]= tail[['long', 'lat']].astype(float)
tail
```

```python
gmaps.configure(api_key=g_key)
```

```python
# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>name</dt><dd>{name}</dd>
<dt>lat</dt><dd>{lat}</dd>
<dt>long</dt><dd>{long}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
yelptop = [info_box_template.format(**row) for index, row in top20.iterrows()]
locations = top20[["lat", "long"]]
# use above variables for layer[]
#markers= gmaps.marker_layer(locations, hotel_info) #info_box_template)
fig = gmaps.figure()

markers= gmaps.marker_layer(locations, info_box_content=yelptop)
fig.add_layer(markers)
fig
```
